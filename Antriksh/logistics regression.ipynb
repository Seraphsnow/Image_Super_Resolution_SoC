```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
def f(x,w,b):
    a = np.matmul(w,np.transpose(x)) + b
    a = 1/(1+np.exp(-a))
    return a
def cost(x,y):
    for m in x:
        if m == 1:
            m = 1-1e-13
        elif m == 0:
            m = 1e-13
    a = -y*(np.log(x)) - (1-y)*(np.log(1-x))
    b = np.sum(a)
    return b
def norm(x):
    mean = np.mean(x,axis = 0)
    std = np.std(x,axis = 0)
    return (x-mean)/std
def boundary(a):
    if a < 0.2:
        a=0
    else :
        a=1
    return a
```

```python
df = pd.read_csv('framingham.csv')
df = df.dropna()
datax = df.to_numpy()
resultx = datax[:, -1] # for last column
datax = datax[:, :-1] # for all but last column
w = np.array([])
for i in range(15):
    w = np.append(w,0)
#training data
data = datax[:3000,:]
data = norm(data)
result = resultx[:3000]
#test data
data_test = datax[-1156:,:]
data_test = norm(data_test)
result_test = resultx[-1156:]
```

```python
b =0
arr1 = f(data,w,b)
initial = 1
final = cost(arr1,result)
l = 0.1
counter = 0
while abs(final - initial) > 0.0000001 :
    w_temp = w - l*np.matmul((f(data,w,b) - result),data)/len(arr1)
    b = b - l*(np.sum(f(data,w,b)-result))/len(arr1)
    w = w_temp
    initial = final
    final = cost(f(data,w,b),result)
    if final > initial :
        l = l/1.1
    counter = counter +1
    print(final)
print(w)
print(b)
print(counter)
```

```python

output = np.matmul(w,np.transpose(data_test))+b
output = 1/(1+np.exp(-output))
print(output)
output = (np.array(list(map(boundary,output))))
print(output)
```

```python
x = output - result_test
correct = 0
false_positive = 0
false_negative = 0
for i in x:
    if i == 0:
        correct+=1
    elif i == -1:
        false_negative +=1
    else:
        false_positive +=1
print("total percentage correctness is",correct/(correct + false_negative+false_positive))
print("total errored results are",np.dot(x,x))
print('accuracy for predicting disease',(sum(result_test)-false_negative)/sum(result_test))
print("accuracy for predicting health",(result_test.shape-sum(result_test)-false_positive)/(result_test.shape-sum(result_test)))

```
