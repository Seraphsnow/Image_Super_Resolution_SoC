```python
import numpy as np
import math
import csv

def p(x_k, x):
    z = np.array([x_k])
    z.astype("float64")
    x.astype("float64")
    return 1/(1+math.exp(-1*(z@x)))


def cost(A, x, y):
    ans = 0
    for i in range(len(y)):
        if int(y[i]) == 1:
            ans += math.log(p(A[i], x))
        else:
            ans += math.log(1 - p(A[i], x))
    return math.fabs(ans)

def del_cost(A, x, y):
    ans = 0
    for i in range(len(A)):
        ans += (y[i] - p(A[i], x))*A[i]
    return ans

y_listed = []
A_listed = []
with open("framingham.csv", "r") as file:
    csvFile = csv.reader(file)
    flag = False
    for lines in csvFile:
        if flag:    
            y_listed.append(lines[-1])
            list1 = [1] + lines[:-1]
            A_listed.append(list1)
        flag = True
x = np.array([.005 for _ in range(16)]).T
A = []
y = []
for i in range(len(A_listed)):
    l1 = []
    if "NA" in A_listed[i]:
        continue
    else:
        y.append(float(y_listed[i]))
        for j in A_listed[i]:
            l1.append(float(j))
    A.append(l1)
y = np.array(y)
y.astype("float64")
A = np.array(A)
learning_rate = 1e-7
cost1 = cost(A, x, y)
print(cost1)
while True:
    x_new = x + learning_rate*del_cost(A, x, y)
    cost_new = cost(A, x_new, y)
    print(cost_new, x_new)
    if cost_new > cost1 and learning_rate >= 1e-15:
        learning_rate /= 2
    elif cost_new > cost1 or cost1 - cost_new < 1e-5:
        break
    else:
        cost1 = cost_new 
        x = x_new
print(x)
```
