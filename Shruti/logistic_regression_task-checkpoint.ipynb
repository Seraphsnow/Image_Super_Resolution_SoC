```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('framingham.csv')
data = data.dropna() #To handle NA values


X = data.iloc[:, :15].values
y = data.iloc[:, 15].values


X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)


X = np.hstack((np.ones((len(y), 1)), X))


def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def cost_function(X, y, theta):
    predictions = sigmoid(X.dot(theta))
    error = -y*np.log(predictions) - (1-y)*np.log(1-predictions)
    cost = np.sum(error) / len(y)
    return cost


def gradient_descent(X, y, theta, alpha, iterations):
    cost_history = []
    for i in range(iterations):
        predictions = sigmoid(X.dot(theta))
        error = predictions - y
        gradient = X.T.dot(error) / len(y)
        theta -= alpha * gradient
        cost = cost_function(X, y, theta)
        cost_history.append(cost)
    return theta, cost_history


theta = np.zeros(X.shape[1])


alpha = 0.1
iterations = 1000

# Run gradient descent
theta, cost_history = gradient_descent(X, y, theta, alpha, iterations)


plt.plot(cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost function over time')
plt.show()


new_data = np.array([[0, 35, 3, 0,0,0,0,0,1,210,140,80,25.34,85,75]])
new_data = (new_data - np.mean(X[:,1:], axis=0)) / np.std(X[:,1:], axis=0)
new_data = np.hstack((np.ones((new_data.shape[0], 1)), new_data))
predictions = sigmoid(new_data.dot(theta))
print(predictions)
```
