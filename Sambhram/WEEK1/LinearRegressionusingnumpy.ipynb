```python
import requests
from bs4 import BeautifulSoup
url = 'https://itc.gymkhana.iitb.ac.in/wncc/soc/'
response = requests.get(url)
if response.status_code == 200:
    content = response.content
    soup = BeautifulSoup(content, 'html.parser')# html parser makes a html tree making it easy to navigate
    Projects = soup.find_all('p', class_ ="lead text-center font-weight-bold text-dark")
    Project = list()
    domain = soup.find_all('div',class_ = 'col-lg-4 col-6 mb-4 shuffle-item') #finding the domain
    domain_name = list()
    
   
    project_len = len(Projects)
    
    
    for P in Projects:
        Project.append(P.text)
        
    for i in range(0,project_len):
        domain_list = eval(domain[i]['data-groups']) #converting into list using eval
        domain_name.append(domain_list[0])
       
else:
    print('Request not granted')
```

```python
Project_divs = soup.find_all('div', class_ = "rounded hover-wrapper pr-3 pl-3 pt-3 pb-3 bg-white")
Pro_links = list()
for Project_link in Project_divs :
    Pro_links.append("https://itc.gymkhana.iitb.ac.in" + Project_link.get('href'))

All_Mentors_list = list()
No_of_Mentees = list()

for i in range(0,len(Pro_links)):  
    url = Pro_links[i]
    response = requests.get(url)
    content = response.content
    soup = BeautifulSoup(content, 'html.parser')
    Mentors = soup.find_all('ul')
    Mentor = Mentors[3] # observation that all the mentors names are present in 4th unordered list
    Mentor = Mentor.find_all('p',class_ = "lead")
    Mentor_name = list()
    Mentees =  soup.find_all('ul')
    
    No_Mentees = Mentees[4] # the fourth ul has no of mentees in all the lists
    No_of_Mentees.append(No_Mentees.text.strip()) #removes /n in the list    
    
    for j in range(0,len(Mentor)):
        Mentor_name.append(Mentor[j].text)
        
    
    All_Mentors_list.append(Mentor_name) #creating a 2d list  

```

```python
import pandas as pd
df = pd.DataFrame({'Project':Project, 'Domain':domain_name, 'Mentors': All_Mentors_list, 'No of Mentees':No_of_Mentees, 'Project_Link': Pro_links})
def remove_brackets(lst): #function for removing the brackets since i was getting brackets
    return ', '.join(lst)
df['Mentors'] = df['Mentors'].apply(remove_brackets)
df.to_csv('WnCC2.csv')
df
```
