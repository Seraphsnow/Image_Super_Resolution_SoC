```python
import numpy as np
import matplotlib.pyplot as plt
def data1():
    x = [np.random.rand() for i in range(1000)]
    y = [x[i] + 0.05*np.random.rand() for i in range(1000)]
    return [x, y]
x = [np.random.rand() for i in range(1000)]
y = [x[i] + 0.05*np.random.rand() for i in range(1000)]
coor = np.array([data1()])

coor= coor.reshape(1000,2) #reshaping according to need
coor1 = coor.reshape(1000,2)
mean = np.mean(coor, axis=0)
centered = np.zeros((1000,2)) #finding the centered dataset

for i in range(1000):
    centered[i]= coor[i] - mean
centered = centered.reshape(2,1000)

covar = np.cov(centered) #the covariance matrix

eigenvalues, eigenvectors = np.linalg.eig(covar)

idx = np.argsort(eigenvalues)[::-1]

eigenvectors = eigenvectors[:,idx] 

eigenvector_max = eigenvectors[:,:1]#finding the eigenvector with maximum eigenvalue

slope = eigenvector_max[1,0]/eigenvector_max[0,0]#finding slope
intercept = np.mean(y) - np.mean(x)*slope
plt.scatter(x,y)
plt.plot(coor1[:,0], slope*coor1[:,0] + intercept, color = 'red'  )
```
