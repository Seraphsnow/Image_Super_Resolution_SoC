```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import math
from scipy.special import expit
```

```python
arr = pd.read_csv('framingham.csv',skiprows = [int(i) for i in range(1000)])
arr = np.array(arr)
shape = arr.shape
```

```python
def f(z):
    return expit(z)
def f_(a, b):
    a = max(a, 1e-15) # add a small constant to avoid taking the log of 0
    a = min(a, 1-1e-15) # subtract a small constant from 1 to avoid taking the log of 1
    return float(-b*(math.log(a)) - (1-b)*(math.log(1-a)))
def div(a, b):
	return float(a/b)
def _f_(a):
	return a*a
def f_des(a):
    if a<=0.5:
        return 0
    else: 
        return 1
```

```python
columns = shape[1]
rows = shape[0]
x_ = arr[:,:columns-1]
std = np.std(x_, axis = 0)
x_ = x_ - np.mean(x_, axis = 0)
x_ = x_/std

one = np.ones(rows).reshape(-1,1)
x_ = np.append(x_, one,1)
y = arr[:,columns-1].reshape(-1)
x = np.copy(x_)
```

```python
w = np.ones(columns)
J_init =10
J_final = 0
delJ = J_init - J_final
eps =  0.000001
alpha = 0.1
J = []
j =0
```

```python
while abs(delJ) > eps:
	z = np.matmul(x,w)
	p = np.array(list(map(f, z)))
	stepx = np.array(list(map(f_, p, y)))
	J_init = np.sum(stepx)

	step = np.matmul(p -y, x)
	w = w - (alpha/rows)*step
	delJ = J_init - J_final
	if J_init > J_final:
		alpha = alpha/1.1
	J_final = J_init
	J.append(J_final)	
	j = j+1	
	print(J_final)
```
